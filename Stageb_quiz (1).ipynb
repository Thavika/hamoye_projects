```python
X = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 42)
```

```python
X_train= X_train.reshape(-1, 1)
y_train= y_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
# Import Metrics for evaluation
from sklearn import metrics
```

```python
# Publish the Metrics for Evaluation
mae=round(metrics.mean_absolute_error(y_test,y_pred),2)
print('Mean Absolute Error (MAE)',mae)
mse=round(metrics.mean_squared_error(y_test,y_pred),2)
print('Mean Square Error (MSE)',mse)
rmse=round( np.sqrt(metrics.mean_squared_error(y_test,y_pred)),2)
print('Root Mean Square Error (RMSE)',rmse)
r2=round(metrics.r2_score(y_test,y_pred),2)
print('R2 score',r2)
```

```python
#Using Ridge Regression with Alpha = 0.5
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.5)
ridge_reg.fit(X_train, y_train)
pred4 = ridge_reg.predict(X_test)           # Use this model to predict the test data
rmse_r=round(np.sqrt(metrics.mean_squared_error(y_test,pred4)),2)
print('Root Mean Square Error (RMSE)',rmse_r ) #RMSE
r2_r=round(metrics.r2_score(y_test,pred4),2)
print('R2 score',r2_r) #R2 Score
```

```python
#Using Ridge Regression with Alpha = 0.4
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(X_train, y_train)
pred6 = ridge_reg.predict(X_test)           
rmse_r1=round(np.sqrt(metrics.mean_squared_error(y_test,pred6)),2)
print('Root Mean Square Error (RMSE)',rmse_r1) #RMSE
r2_r1=round(metrics.r2_score(y_test,pred6),2)
print('R2 score',r2_r1) #R2 Score
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(X_train, y_train)
pred5=lasso_reg.predict(X_test)# Use this model to predict the test data
rmse_l=round(np.sqrt(metrics.mean_squared_error(y_test,pred5)),3)
print(rmse_l)
print('Root Mean Square Error (RMSE)',np.sqrt(metrics.mean_squared_error(y_test,pred5))) #RMSE
r2_l=round(metrics.r2_score(y_test,pred5),2)
print('R2 score',r2_l ) #R2 Score
```

```python
#comparing the effects of regularisation
def get_weights_df(model, feat, col_name):            
 #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df
```

```python
linear_model_weights = get_weights_df(regressor, X_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, X_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, X_train, 'Lasso_weight')
final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
print(final_weights)
```
